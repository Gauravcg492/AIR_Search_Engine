```python
# gives everything except zones list
new = df.groupby(['token','doc-id']).agg(pd.Series.tolist)
new
```

```python
import math 
import json
count =0
it=0
inverted_index = {}
N = 94858 #hardcoded because static corpus

for index, row in new.iterrows():
    # TEST COMMENT BEGINS HERE
    # if(count==200):
    #     break
    # count+=1
    # TEST COMMENT ENDS HERE
    
    token = index[0]
    doc_id = index[1]
    pos_list = row['pos']
    zone_list = row['zone']
    
    tf = len(pos_list)
    w = 1 + math.log(tf,10)
    tuple_entry = (doc_id, pos_list, w) #for each doc  
    
    if(token not in inverted_index.keys()):
        idf =0.0 #calculate all idfs in the end
        inverted_index[token] =[idf, [], [], set() ]

    inverted_index[token][1].append(tuple_entry)
    inverted_index[token][3].update(zone_list)

#idf calculation
for key in inverted_index.keys():
    dft = len(inverted_index[key][1])
    inverted_index[key][0] = math.log(N/dft, 10) #base 10 

#convert all zone "sets" to lists
for key in inverted_index.keys():
    inverted_index[key][3] = list(inverted_index[key][3])

#construct champion list:
for key in inverted_index.keys():
    #sort each posting list in descending order of "tf"
    sorted_index = sorted(inverted_index[key][1],key = lambda x: x[2], reverse= True) 
    #top 20 elements considered to be part of champion list
    inverted_index[key][2] = sorted_index[:100] 
    
#write dictionary to a new file:
with open("../../inv_index.json", "w") as outfile:
    #for key in inverted_index.keys():
        #new_dict= {}
        #new_dict[key] = inverted_index[key]
    json.dump(inverted_index, outfile)
    outfile.write("\n")
        



 
```
